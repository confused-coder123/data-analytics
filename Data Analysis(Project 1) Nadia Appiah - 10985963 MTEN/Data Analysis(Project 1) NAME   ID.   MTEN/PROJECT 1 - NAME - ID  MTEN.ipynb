## Statistical inferences of individual columns

```python
new_data = data.rename(columns = {'fixed acidity':'fixed_acidity' , 'volatile acidity':'volatile_acidity','citric acid':'citric_acid','residual sugar':'residual_sugar','free sulfur dioxide':'free_sulfur_dioxide','total sulfur dioxide':'total_sulfur_dioxide'})
```

### Mean of individual columns

```python
mean_fd = new_data.fixed_acidity.mean()
mean_vd = new_data.volatile_acidity.mean()
mean_cd = new_data.citric_acid.mean()
mean_rs = new_data.residual_sugar.mean()
mean_cl = new_data.chlorides.mean()
mean_fsd = new_data.free_sulfur_dioxide.mean()
mean_tsd = new_data.total_sulfur_dioxide.mean()
mean_ds = new_data.density.mean()
mean_pH = new_data.sulphates.mean()
mean_al = new_data.alcohol.mean()
mean_ql = new_data.quality.mean()
```

```python
print(f'fixed_acidity: {mean_fd}')
print(f'volatile_acidity: {mean_vd}')
print(f'citric_acid: {mean_cd}')
print(f'residual_sugar: {mean_rs}')
print(f'chlorides: {mean_cl}')
print(f'free_sulfur_dioxide: {mean_fsd}')
print(f'total_sulfur_dioxide: {mean_tsd}')
print(f'density: {mean_ds}')
print(f'pH: {mean_pH}')
print(f'alcohol: {mean_al}')
print(f'quality: {mean_ql}')
```
